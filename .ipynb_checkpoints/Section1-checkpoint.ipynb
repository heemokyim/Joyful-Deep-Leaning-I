### Problem 1: Closed-Form Maximum Likelihood

Assume that we are given $n$ iid samples $(x_1, ..., x_n)$ from each $P(x \ | \ \theta)$ below. Compute the maximum likelihood estimates (MLEs) for the parameter $\theta$ of given distributions. 

**Q1.1**: $P(x \ | \ \theta) = \theta e^{-\theta x}$ for $x \geq 0$

**Q1.2**: $P(x \ | \ \theta) = \frac{1}{\theta}$ for $ 0 \leq x \leq \theta$

**A1.1**: $\log P(x|\theta) = \log(\theta) - \theta x$. Then the log-likelihood for the samples is $L(\theta) = n \log(\theta) + \theta(\sum_{i=1}^n x_i)$. Taking derivative of the log-likelihood function gives $\frac{\partial L(\theta)}{\partial \theta} = n/\theta - \theta \sum_{i=1}^n x_i$. Hence, using the first order optimality gives $\hat \theta = \frac{n}{\sum_{i=1}^n x_i}$.

Next, the second derivative of the log-likelihood at $\hat \theta$ is $-(\sum_{i=1}^n x_i)^2/n - \sum_{i=1}^n x_i$, which is negative due to the fact $x_i > 0$. Hence $\hat \theta$ is indeed a maximum.

** OR **
$$
P(x_1,x_2,\dots,x_n \ | \ \theta)=\prod_{i=1}^{n}P(x_i \ | \ \theta)
$$

So the log-likelihood is

$$
\log P(x_1,x_2,\dots,x_n \ | \ \theta)=\sum_{i=1}^{n}\log P(x_i \ | \ \theta)=\sum_{i=1}^{n}(\log\theta-\theta x_i)=n\log\theta-\theta\sum_{i=1}^{n}x_i
$$

The MLE for $\theta$ is:

$$
\hat{\theta}_{ML}=\text{arg}\max_{\theta}\log P(x_1,x_2,\dots,x_n \ | \ \theta)
$$

By setting the derivative of $\log P(x_1,x_2,\dots,x_n \ | \ \theta)$ with respect to $\theta$ to zero, we get 

$$
n\frac{1}{\theta}-\sum_{i=1}^{n}x_i=0
$$

and therefore

$$
\hat{\theta}_{ML}=\frac{n}{\sum_{i=1}^{n}x_i}.
$$


**A1.2**: 

The log-likelihood function for the samples is $L(\theta) = (1/\theta)^n$ if $\theta \geq \max\{x_1,\, x_2,\, \cdots,\, x_n\}$ and $L(\theta) = 0$ if $\theta < \max\{x_1,\, x_2,\, \cdots,\, x_n\}$. Hence when $\hat \theta = \max\{x_1,\, x_2,\, \cdots,\, x_n\}$, the likelihood function reaches its maximum.

** OR **

$$
P(x_1,x_2,\dots,x_n \ | \ \theta)=\prod_{i=1}^{n}P(x_i \ | \ \theta)=\frac{1}{\theta^n}
$$

So the log-likelihood is

$$
\log P(x_1,x_2,\dots,x_n \ | \ \theta)=-n\log\theta
$$

and

$$
\frac{\partial \log P(x_1,x_2,\dots,x_n \ | \ \theta)}{\partial \theta}=-\frac{n}{\theta}.
$$

The log-likelihood decreases as $\theta$ increases. Since we have obtained samples $x_1,\dots,x_n$, then $\theta$ should be large enough so that all samples lie in the interval $[0,\theta]$. So the MLE for $\theta$ is:

$$
\hat{\theta}_{ML}=\text{arg}\max_{\theta}\log P(x_1,x_2,\dots,x_n \ | \ \theta)=\max\{x_1,x_2,\dots,x_n\}.
$$

**OR**

By the same way with Ans1.1; 
$$\theta_{MLE} = \arg \max_{\theta} \frac{1}{\theta^{n}} = \arg \max_{\theta} -n \log \theta \text{ where -n $\log \theta$ is a monotonically decreasing function} \\ $$ 
So, the function has its maximum value at its minimum $\theta$ value in its domain. But we are given $0 \leq x \leq \theta$ so the minimum value that $\theta$ can take is $\theta_{MLE} = \max (x_{1},...,x_{n})$ 
The answer makes sense since we try to narrow the domain of the probability density as much as possible to increase the likelihood in our case.

### Problem 2: Gradient for Maximum Likelihood Estimation

Sometimes, we can't find MLEs in closed-form, in this case, we can use an iterative scheme to solve resulting optimization problem $\theta_{MLE} = \arg\max P(x \ | \ \theta)$. First-order iterative schemes (e.g. Gradient Descent, Stochastic Gradient Descent, Nesterov Accelerated Gradient Descent) require gradient information at each visited point. Compute gradients for log-likelihood of the following model:

- we have $X \in \mathbf R^{n \times k}$ - constant data matrix, $\mathbf x_i$ - vector corresponding to a single data point
- $\theta$ is a $k$-dimensional (unknown) weight vector
- $\varepsilon \sim \text{Student}(v)$ is a $n$-dimensional (unknown) noise vector
- and we observe vector $\mathbf y = X\theta + \varepsilon$

Which means that
$$ P(y_i \ | \ \mathbf x_i, \theta, v) = \frac{1}{Z(v)} \Big(1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v}\Big)^{-\frac{v+1}{2}}$$

where $Z$ stands for a normalization constant that does not depend on $\theta$, and $v$ is a constant parameter.

Denote log-likelihood as $\mathcal L(\theta)$ and log-likelihood for a single sample $i$ as $\mathcal L_i(\theta) = \mathcal L(\theta \ | \ \mathbf x_i, y_i)$.

**Q2.1**: Find $\frac{\partial \mathcal L_i(\theta)}{\partial \theta}$.

**A2.1**: 

First find the log-likelihood function. $\mathcal L_i (\theta) = \log (P(y_i|x_i, \theta, v)) = C -\frac{v+1}2 \log(1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v})$. Then 

$$\frac{\partial \mathcal L_i(\theta)}{\partial \theta} = \frac{\partial}{\partial \theta} -\frac{v+1}2 \log(1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v}) = -\frac{v+1}2 \frac{-2(y_i - \theta^T \mathbf x_i)}{1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v}} \mathbf x_i = \frac{\left(v+1\right)\, \left(y_i - \theta^T \mathbf x_i \right)}{v \left( 1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v} \right)} \mathbf x_i $$

** OR **

$$L_i(\theta) = \mathcal L(\theta \ | \ \mathbf x_i, y_i) = - \log(Z(v)) - \frac{v+1}{2} \log \Big(1 + \frac{(\theta^T \mathbf x_i - y_i)^2}{v}\Big)$$
$$\frac{\partial \mathcal L_i(\theta)}{\partial \theta} = -(v+1) \frac{\frac{\theta^{T}x_{i} - y_{i}}{v} \cdot x_{i}}{1 + \frac{(\theta^{T}x_{i} - y_{i})^{2}}{v}} = -(v+1) \frac{(\theta^{T}x_{i} - y_{i}) \cdot x_{i}}{v + (\theta^{T}x_{i} - y_{i})^{2} }$$

**Q2.2**: Check the result obtained in the previous section by coding an SGD for $\mathcal L(\theta)$ and plotting how function value changes as a function of a number of passed gradient steps. Try multiple learning rates (1e-1, 1e-3, 1e-6), explain the result. Consider $Z(v)=1$ and $v=4$ for plotting.

We marked places where you are expected to add/change your own code with **`## -- ! code required`** comment.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

def generate_data(n=100, k=10, v=4):
    X = np.random.rand(n, k)*2-1
    theta = np.random.rand(k, 1)
    noise = np.random.standard_t(4, n).reshape((-1, 1))
    return X, X @ theta + noise

def loss(X, y, theta):
    res = np.dot(X, theta) - y
    logits = np.log(t.pdf(res, df=4))
    loss = -np.mean(logits)
    return loss #negative log-likelihood

## OR 

def loss(X, y, theta):
    v=4
    log_loss= (v+1)/2*(np.log(1 + (X @ theta-y)**2/v))
    return np.sum(log_loss)

def loss_grad_i(X, y, theta, i):
    v = 4
    const = (v+1)/(1.0*v)
    res = y[i] - np.dot(X[i], theta) 
    demon = np.square(res)/(1.0*v) + 1
    weight = const*res/demon
    loss_grad_i = -weight * X[i] #negative LL, negative sign here
    return loss_grad_i.reshape(np.shape(theta))

def sgd_plot(theta_init, X, y, n_steps=100):
    for learning_rate in [1e-1, 1e-3, 1e-6]:
        theta = theta_init[:]
        losses = []
        for step_n in range(n_steps):
            for i in range(X.shape[0]):
                # References: https://en.wikipedia.org/wiki/Stochastic_gradient_descent
                theta = theta - learning_rate * loss_grad_i(X, y, theta, i) ## ! -- code required
                loss_i = loss(X, y, theta) ## ! -- code required
                losses.append(loss_i)
        plt.plot(losses)
        plt.show()
    
def main(n, k, v):
    X, y = generate_data(n, k, v)
    theta_init = np.random.rand(k, 1)
    sgd_plot(theta_init, X, y)
```

```python
main(100, 2, 4)
```

As can be seen from the plots, when the learning rate is too large, it starts to oscillate when we are close to minima. If the learning rate is too small, then the steps become too small and it takes a lot of time to converge. For our problem, $10^{-3}$ is a proper learning rate

### Problem 3: Matrix Derivatives

**Q3.1: Multivariate Gaussian **

Assume that our data is distributed according to [multivariate Gaussian](https://en.wikipedia.org/wiki/Multivariate_normal_distribution#Likelihood_function) with $\bar \mu$ mean and $\Sigma$ covariance matrix: $$(\mathbf x_1, \dots, \mathbf x_n) \sim \mathcal N(\bar \mu, \Sigma).$$ 

Using rules of [matrix derivatives](http://www2.imm.dtu.dk/pubdb/views/edoc_download.php/3274/pdf/imm3274.pdf) [57, 59], get a $\frac{\partial \mathcal L(\theta)}{\partial \Sigma}$ in matrix form.

**A3.1**:
$$ \mathcal L(\Sigma) = - \frac12 \left( \ln \left(|\Sigma|\right) + (x- \bar \mu)^T \Sigma^{-1} (x-\bar \mu) + k \ln\left(2\pi\right) \right)$$ 

Hence
$$ \frac{\partial \mathcal L(\Sigma)}{\partial \Sigma} = -\frac12 \left( \frac{1}{|\Sigma|} |\Sigma| \Sigma^{-T}  - \Sigma^{-T} (x- \bar \mu)(x-\bar \mu)^T\Sigma^{-T} \right) =  -\frac12 \left(\Sigma^{-T}  - \Sigma^{-T} (x- \bar \mu)(x-\bar \mu)^T\Sigma^{-T} \right)$$

** OR **

theta represents the variables, so it corresponds to $\bar \mu, \Sigma$ in this problem.
$$ P(x_{1},...x_{n} \ | \ \bar \mu, \Sigma) = \prod_{i=1}^{n} P(x_{i} \ | \ \bar \mu, \Sigma) = \frac{1}{\sqrt{(2 \pi^{nk} |\Sigma |^{n})}} exp \Big( - \frac{1}{2} \sum_{i=1}^{n} (x_{i}-\bar \mu) ^{T} \Sigma^{-1} (x_{i}-\bar \mu)\Big) $$
$$ \mathcal L(\theta) = \frac{-1}{2} \sum_{i=1}^{n} (x_{i}-\bar \mu) ^{T} \Sigma^{-1} (x_{i}-\bar \mu) - \frac{nk}{2} \log{(2 \pi)} - \frac{n}{2} \log{|\Sigma|} $$

by using the mentioned formulas:
$$\frac{\partial \log{|\Sigma|}}{\partial \Sigma} = (\Sigma^{-1})^{T} = \Sigma^{-1} \text{(eqn 57) since $\Sigma$ is symmetric } $$
$$\text{follows from eqn 59; }\frac{\partial a^TX^{-1}b}{\partial X} = -X^{-1}ab^TX^{-1} $$
$$a= x_{i}- \mu \text{ and } b =(x_{i}- \mu)^{T} \text{gives us the direct answer of } \frac {\sum_{i=1}^{n} \partial \Big(  (x_{i}-\bar \mu) ^{T} \Sigma^{-1} (x_{i}-\bar \mu) \Big) }{\partial \Sigma}= - \Sigma^{-1}(x_{i}-\bar \mu)  (x_{i}-\bar \mu)^{T}  \Sigma^{-1} $$

$$  \text{So the result is:} \ \frac{\partial \mathcal L(\theta)}{\partial \Sigma}= \Sigma^{-1} - \Sigma^{-1}(x_{i}-\bar \mu) (x_{i}-\bar \mu)^{T}  \Sigma^{-1} $$

**Q3.2: Multi-target Linear Regression**
- we have $X \in \mathbf R^{n \times k}$ is a constant data matrix
- $\theta$ is a $k \times m$-dimentional **weight matrix**
- $\varepsilon_{ij} \sim \mathcal N(0, \sigma_\epsilon)$ is a normal noise ($i \in [0; n], j \in [0;m]$)
- and we observe a matrix $Y = X\theta + \varepsilon \in \mathbf R^{n \times m}$

$$\varepsilon = Y - X\theta \sim \mathcal N_n(0, \sigma_\epsilon I)$$

$$\mathcal L(\theta) = \log P(Y - X\theta \ | \ \theta) = \log \mathcal N_n(Y - X\theta \ | \ 0, \sigma_\epsilon I)$$

$$\theta_{MLE} = \arg \max_{\theta} \mathcal L(\theta) = \arg \min_{\theta} \text{loss}(\theta) = \arg \min_{\theta} \big( ||Y-X\theta||^2_F \big)$$

In equations above $I$ stands for an identity matrix. **Using rules of [matrix derivatives](http://www2.imm.dtu.dk/pubdb/views/edoc_download.php/3274/pdf/imm3274.pdf)** [137, 132], show for an MLE loss:  $\text{loss}(\theta) = ||Y-X\theta||^2_F$ that:

**Q3.2.1:** $\frac{\partial\text{loss}(\theta)}{\partial \theta} = -2X^T (Y-X\theta)$

**Q3.2.2:** $\theta_{MLE} = (X^T X)^{-1} X^T Y$

**Hint:** in our case [see Matrix Cookbook, eq. 137], $g(U) = ||U||^2_F$ - squared Frobenius norm and $U(\theta) = f(\theta) = Y - \theta X$ - linear mapping.

**Note:** That is a multi-target problem, so $\theta$ is a matrix, so you have to take the derivative wrt matrix.

**A3.2.1:**

Denote $U(\theta) = Y-X\theta$. Then the MLE loss is $\| U \|_F^2$.

So it is necessary to proof that $\frac{\partial\text{loss}(\theta)}{\partial \theta_{i,j}} = -[2X^T (Y-X\theta)]_{i,j} = -X_i^T [2(Y-X\theta)]_j $. Here the single subscript $M_i$ means the $i$th column of matrix $M$.

$\frac{\partial \|U\|_F^2}{\partial U} = 2U = 2(Y-X\theta)$. And $\frac{\partial U}{\partial \theta_{ij}}$ is a $n \times m$ matrix with $j$'th column $-X_i$ and rest of elements zeros. Therefore,
$$ \frac{\partial\text{loss}(\theta)}{\partial \theta_{i,j}} = \mathrm{tr} \left( \frac{\partial \|U\|_F^2}{\partial U}  \frac{\partial U}{\partial \theta_{ij}} \right) = -([2(Y-X\theta)]_j)^T X_i = -X_i^T [2(Y-X\theta)]_j$$
Hence we finish the proof.

** OR **

$$loss(\theta) = \big( ||Y-X\theta||^2_F \big)$$
Firstly we use eqn 137 and take $g(\cdot)=  || \cdot ||^2_F $ and $U=Y-\theta X$
Then eqn 132 allows us to compute first term in eqn 137 easily. 
$$\Big( \frac{\partial g(U)}{\partial U} \Big)^{T} = \Big( \frac{\big( ||Y-X\theta||^2_F \big)}{\partial (Y-\theta X)} \Big)^{T} = 2(Y-X \theta)^{T}$$
By eqn 137,
$$ \Big( \frac{\partial ||Y-X\theta||^2_F}{\partial \theta_{ij}} \Big)^{T} = Tr\Big(2(Y-X \theta)^T \frac{\partial (Y-\theta X)}{\partial \theta_{ij}}  \Big) $$

$$\frac{\partial (Y-\theta X)}{\partial \theta_{ij}}=$$
\begin{bmatrix}
    0       & 0 & X_{1i} & \dots & 0 \\
    0       & 0 & X_{2i} & \dots & 0 \\
             &\dots     &\dots       &\dots \\
    0       & 0 & X_{ni} & \dots & 0
\end{bmatrix}

$$ \text{So} \  Tr(2(Y-X \theta)^T \frac{\partial (Y-\theta X)}{\partial \theta_{ij}})= ((Y-X \theta)^T X_{:i})_{j} = (Y-X \theta)^T_{j:}X_{:i}= X_{i:}^T(Y-X \theta)_{:j} \text{where $()_{j}$ represents the jth element}$$
Then bring it to the matrix form:
$$ \frac{\partial \big( ||Y-X\theta||^2_F \big)}{\partial \theta} = X^T(Y-X \theta) $$ 

**OR**

By the chain rule of matrix derivatives [Eq. 137], 
$$
\frac{\partial g(U)}{\partial \theta_{ij}}=\text{Tr}\Big[\Big(\frac{\partial g(U)}{\partial U}\Big)^{\text{T}}\frac{\partial U}{\partial \theta_{ij}}\Big]
$$
$$
=\text{Tr}\Big[2U^{\text{T}}\frac{\partial U}{\partial \theta_{ij}}\Big]
$$
$$
=\text{Tr}\Big[2(Y-X\theta)^{\text{T}}\frac{\partial (Y-X\theta)}{\partial \theta_{ij}}\Big]
$$
$$
=\text{Tr}\Big[-2(Y-X\theta)^{\text{T}}\Big(X\frac{\partial \theta}{\partial \theta_{ij}}+\frac{\partial X}{\partial \theta_{ij}}\theta\Big)\Big]
$$
$$
=\text{Tr}\Big[-2(Y-X\theta)^{\text{T}}X\frac{\partial \theta}{\partial \theta_{ij}}\Big]
$$
$$
=\text{Tr}\Big[-2\Big(\frac{\partial \theta}{\partial \theta_{ij}}\Big)^{\text{T}}X^{\text{T}}(Y-X\theta)\Big]
$$

Therefore,
$$
\frac{\partial g(U)}{\partial \theta}=-2X^{\text{T}}(Y-X\theta).
$$

**A3.2.2:**
Using the first order of optimality, $\frac{\partial\text{loss}(\theta)}{\partial \theta}|_{\theta = \theta_{MLE}}=0$, i.e. $- 2 X^T (Y-X\theta_{MLE}) = 0$. Solving this equality yields $\theta_{MLE} = (X^T X)^{-1} X^T Y$.

Next I validate this solution is indeed a minimum. The Hessian of the loss function is $\frac{\partial^2 \text{loss}(\theta)}{\partial \theta^2} = \frac{\partial}{\partial \theta}\left(- 2 X^T (Y-X\theta) \right)= X^T X \geq 0$ for all $\theta$. Hence the solution is a minimum.

### Problem 4: Logistic Regression

Assume that we observe data generated via the following model:

- $X \in \mathbf R^{n \times k}$ - constant data matrix
- $\theta_j \sim \mathcal N(0, \sigma_\theta)$; $\theta$ is $k$-dimensional random vector

- and we observe $\mathbf y = \mathbf 1[X \theta + \varepsilon \geq 0]$ - boolean random vector

or, in other words: there exists some data $X$, we sample weights vector $\theta$ of the corresponding shape from multivariate normal distribution, multiply features from $X$ by weights from $\theta$, add random logistically distributed noise to resulting numbers, and observe whether resulting (noisy) random variable exceeds zero. So we are given the data matrix X and y, we want to estimate the weight vector $\theta$ as accurately as possible.

\end{align}

Proof : 

Case 1: Assume $y_{i}=1$. So by the claimed equation $P(y_{i} =1 \ | \ \theta, x_{i})$ =  $\text{logistic}(\theta^T \mathbf x_i) $   which coincides with our result in Q4.1, after replacing $\sigma_{\varepsilon}=1$

Case 2: Assume $y_{i}=0$ So by putting it to the equation, $P(y_{i} =0 \ | \ \theta, x_{i})$ =  1 - $\text{logistic}(\theta^T \mathbf x_i)$ which is also true since $\sum_{y_{i} \in {0,1}} P(y_i \ | \ \theta, \mathbf x_i) = 1$ 

**Q4.3**: $ \log P(y_i \ | \ \theta, \mathbf x_i) = y_i \theta^T \mathbf x_i - \log(1 + \exp(\theta^T \mathbf x_i))$

**A4.3:**

$$ \log P(y_i \ | \ \theta, \mathbf x_i) = \log \big( (\text{logistic}(\theta^T \mathbf x_i))^{y_i} (1-\text{logistic}(\theta^T \mathbf x_i))^{(1-y_i)} \big) \\ = y \log( \exp( \theta^T \mathbf x_i)) - y \log(1 + \exp(\theta^T \mathbf x_i)) -(1-y) \log(1 + \exp(\theta^T \mathbf x_i)) \\ = y_i \theta^T \mathbf x_i - \log(1 + \exp(\theta^T \mathbf x_i))$$

The vector $\mathbf 1_{n \times 1}$ denotes column of ones.

**Q4.4**: $ \mathcal L_{MLE}(\theta) = \log P(\mathbf y \ | \ X, \theta) = y^T X \theta - \log(1 + \exp(X \theta)) \cdot \mathbf 1_{n \times 1}$

**A4.4**: 

$$ \mathcal L_{MLE}(\theta) = \log P(\mathbf y \ | \ X, \theta) = \log \big( P(y_1 \ | \ \theta, \mathbf x_1) P(y_2 \ | \ \theta, \mathbf x_2) \cdots P(y_n \ | \ \theta, \mathbf x_n) \big) \\= \sum_{i=1}^{n} \log P(y_i \ | \ \theta, \mathbf x_i) = \sum_{i=1}^{n} y_i \theta^T \mathbf x_i - \log(1 + \exp(\theta^T \mathbf x_i))= y^T X \theta - \log(1 + \exp(X \theta)) \cdot \mathbf 1_{n \times 1}$$

One way of estimating true weight vector $\theta$ is by finding an MLE estimate. To make use of prior information,  one might want to find [maximum a posteriori estimation](https://en.wikipedia.org/wiki/Maximum_a_posteriori_estimation) (MAP) that can be derived from likelihood $P(X|\theta)$ and prior $P(\theta)$ over weights as following (Bayes rule):

$$ \theta_{MAP} = \arg\max_\theta P(\theta | X) = \arg\max_\theta \frac{P(X|\theta)P(\theta)}{\int_{\Theta} P(X|\theta)P(\theta) d \theta} = \arg\max_\theta \big[ \log P(X|\theta) + \log P(\theta) - \log \int_{\Theta} P(X|\theta)P(\theta) d \theta \big] \\ = \arg\max_\theta \big[ \log P(X|\theta) + \log P(\theta) \big] $$

In the model definition above, normal prior over weights is specified by line $\theta_j \sim \mathcal N(0, \sigma_\theta)$, which essentially means that "weight vector components should not be too far from zero".

**Q4.5**: $\mathcal L_{MAP}(\theta) = \log P(\theta \ | \ X, \mathbf y) = y^T X \theta - \log(1 + \exp(X \theta)) \cdot \mathbf 1_{n \times 1} - \lambda \ ||\theta||^2_2 +C $, $\lambda = \frac{1}{2\sigma_\theta}$

**A4.5**: Since $\mathcal L_{MAP}(\theta) = \log P(X|\theta) + \log P(\theta)$, it is sufficient to show $\log P(\theta) = - \lambda \ ||\theta||^2_2 +C$ according to the result of Q4.4. 

$$ \log P(\theta) = \log\big( (2\pi \sigma_\theta )^{-k/2} \exp(\frac{\|\theta\|_2^2}{2\sigma_\theta}) \big) = C + \lambda \ ||\theta||^2_2$$

**OR**

$$\theta_{MAP} = \arg\max_\theta P(\theta | X,y) = \arg\max_\theta (\log P(y \ | \ X, \theta)+ \log P(\theta \ | \ X)) $$
We know the first term from 4.4, so we need to find the latter term.

\begin{align}
\log P(\theta \ | \ X) &= \log P(\theta) \sim \mathcal N(0, \sigma_\theta) \ \text{since X is a constant matrix} \\
\log P(\theta) &= \sum_{j=1}^{n} \log p(\theta_{j}^2) = \sum_{j=1}^{n} \log \frac{1}{\sqrt{2 \pi \sigma_{\theta}}} e^{- \frac{ \theta_{j}^2}{2 \sigma_{\theta} ^2}} = \sum_{j=1}^{n} - \frac{ \theta_{j}^2}{2 \sigma_{\theta} ^2} + c \ \text{where $c=\frac{-1}{2} \log 2 \pi \sigma_{\theta} $ which is independent of $\theta$} \\
&= c - \frac{\sum_{j=1}^{n}  \theta_{j}^2}{2 \sigma_{\theta}^2} = c - \lambda \ ||\theta||^2_2 \text{where $\lambda = \frac{1}{2\sigma_\theta^2}$} \\
\text{So} \  \mathcal L_{MAP}(\theta) &= \log P(y \ | \ X, \theta)+ \log P(\theta ) = y^T X \theta - \mathbf 1_{n \times 1}^T \cdot \log(\mathbf 1_{n \times 1} + \exp(X \theta)) - \lambda \ ||\theta||^2_2 + C
\end{align}

**Q4.6**: $\frac{\partial \mathcal L_{MAP}(\theta)}{\partial \theta} = X^T(y - \text{logistic}(X\theta))) - 2 \lambda \theta $

**A4.6:** 

$\frac{\partial}{\partial \theta} (y^T X \theta) = X^T y$. $\frac{\partial}{\partial \theta} (- \lambda \ ||\theta||^2_2) = - 2 \lambda \theta$. And it is sufficient to show that $\frac{\partial}{\partial \theta} \log(1 + \exp(X \theta)) \cdot \mathbf 1_{n \times 1} = X^T \text{logistic}(X\theta)$

$$ \frac{\partial}{\partial \theta} \log(1 + \exp(X \theta)) \cdot \mathbf 1_{n \times 1} = \frac{\partial}{\partial \theta} \sum_{i=1}^n \log(1 + \exp(\mathbf x_i^T \theta)) = \sum_{i=1}^n \text{Logstic}(\mathbf x_i^T \theta) x_i^T = X^T \text{logistic}(X\theta)$$

**OR**

$$\frac{\partial \mathcal (y^T X \theta)}{\partial \theta} = \frac{\partial \mathcal ((X^T y)^T \theta)}{\partial \theta} = X^T y $$

$$ \mathbf 1_{n \times 1}^T \cdot \log(\mathbf 1_{n \times 1} + \exp(X \theta)) = \sum_{i=1}^{n} \log (1 + exp(x_{i}^T \theta))
$$

$$\frac{\partial( \mathbf 1_{n \times 1}^T \cdot \log(\mathbf 1_{n \times 1} + \exp(X \theta))}{\partial \theta} = \frac{\partial (\sum_{i=1}^{n} \log (1 + exp(x_{i}^T \theta)) }{\partial \theta} = \sum_{i=1}^{n} \frac{\partial \log (1 + exp(x_{i}^T \theta))}{\partial \theta}$$
$$ \text{Label} \  D_{\theta_{j}} = \sum_{i=1}^{n} \frac{\partial \log (1 + exp(x_{i}^T \theta))}{\partial \theta_{j}} \text{which represents the partial derivatives for each entry j=1,...,k} $$
$$ D_{\theta_{j}} = \sum_{i=1}^{n} \frac{exp(x_{i}^T \theta)}{1 + exp(x_{i}^T \theta)}  x_{ij} = \sum_{i=1}^{n} x_{ij} \text{logistic}(x_{i}^T \theta) \ \text{for j=1,...,k}$$ 

$$\text{so when we write it in matrix form ; } \ \frac{\partial( \mathbf 1_{n \times 1}^T \cdot \log(\mathbf 1_{n \times 1} + \exp(X \theta))}{\partial \theta} =X^T \text{logistic} (X \theta)$$
$$ \frac{\partial (\lambda \ ||\theta||^2_2)}{\partial \theta} = 2 \lambda \theta \ \text{we used this derivative formula before} $$
 $$  \text{Collecting all terms give:} \   \frac{\partial \mathcal L_{MAP}(\theta)}{\partial \theta} = X^T(y - \text{logistic}(X\theta))) - 2 \lambda \theta $$

The prior part of the likelihood ($||\theta||_2^2$) may also be interpreted as [Tikhonov regularizer](https://en.wikipedia.org/wiki/Tikhonov_regularization). When the problem is ill-posed, it might have multiple solutions, and regularizer lets us define which of them is more preferable ("with the smaller L2 norm - in our case") by modifying loss, and as a result, letting an iterative scheme converge better.

**Q4.7**: What does $\lambda$ stand for?

**A4.7:** That is also a regularization paratemter. It defines the prior over weights: large $\lambda$ leads to small $\text{Var}(\theta)$ and vice versa. This means that regularization restricts variance of weights (if we want them all to be centered around zero or not).

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_logistic_regression_data(n_samples=200, k_featues=2, sigma_eps=0.1, 
                                      sigma_theta=1, feature_correlation_rate=0, train_test_ratio=4):
    assert 0 <= feature_correlation_rate <= 1
    X_noncorr = np.random.rand(n_samples, k_featues)*2-1 # [0, 1] -> [-1, 1] center around zero
    mixture_matrix = np.eye(k_featues)*(1-feature_correlation_rate) + feature_correlation_rate/k_featues
    X = X_noncorr @ mixture_matrix
    theta_real = np.random.normal(scale=sigma_theta, size=k_featues).reshape((-1, 1)) # column
    noise = np.random.logistic(scale=sigma_eps, size=n_samples).reshape((-1, 1))
    y = ((X @ theta_real + noise) > 0).astype(np.int16).reshape((-1, 1))
    split_idx = int(n_samples * train_test_ratio / (train_test_ratio + 1))
    return X[:split_idx, :], y[:split_idx], X[split_idx:, :], y[split_idx:], theta_real

def plot_2d_data(subfig, X, y, X_test, y_test, theta):
    assert len(X.shape) == 2
    subfig.scatter(X[:, 0], X[:, 1], color = [['r', 'b'][c] for c in np.ravel(y)])
    point_color_arr = [['r', 'b'][c] for c in np.ravel(y_test)]
    subfig.scatter(X_test[0], X_test[1], marker='x', color = point_color_arr)
    subfig.plot((1, -1), (-theta[0]/theta[1], theta[0]/theta[1]))
    subfig.set_xlim(-1, 1)
    subfig.set_ylim(-1, 1)
    
def subfig_iter(it, val_name=None, n_col=2, base_size=(10,2), title=None, fmt='%s'):
    n_rows = int(len(it)/n_col) + len(it) % n_col
    size = (base_size[0], base_size[1]*len(it))
    fig, _axes = plt.subplots(n_rows, n_col, figsize=size)
    for subfig, val in zip(np.ravel(_axes).tolist(), it): 
        yield (val, subfig)
        if val_name is not None:
            if val_name == '':
                subtitle = fmt % (val)
            else:
                subtitle = ('%s = '+fmt) % (val_name, val)
            subfig.set_title(subtitle)
            subfig.legend(loc='best')
    if title is not None:
        fig.suptitle(title, verticalalignment='top') 

def test_draw():
    data_args_dicts = [
        dict(), {'sigma_eps': 0.2}, {'sigma_eps': 0.5}, {'feature_correlation_rate': 0.3}, 
        {'feature_correlation_rate': 0.7}, {'feature_correlation_rate': 0.9},
    ]
    for arg_dict, subfig in subfig_iter(data_args_dicts, val_name=''):
        data = generate_logistic_regression_data(**arg_dict)
        X_train, y_train, X_test, y_test, theta_real = data
        plt.title(str(arg_dict))
        plot_2d_data(subfig, X_train, y_train, X_test, y_test, theta_real)
        
    plt.show()

test_draw()
```

**A5.1.1:** For $\alpha = 0 $: Matrix M is identity matrix. MX gives us the initial features separately, with no correlation between them (if they are initially independent)

**Q5.2**: Why did we implement $\text{softplus}(x) = \log(1+\exp(x))$ as shown below? What is numerical under- and overflow and how people deal with it?

**A5.2**

Overflow in numerical computations occures if we try to store a floating point value that is larger then maximum number that might be stored by hardware. Underflow happens when we try to work with numbers of different magnitudes, that leads to loss of precision, that might lead to numerical instability. For example, `1e10+1e-10 == 1e10`. In our case, if we compute `softplus` naively, we might lose precision duing `x + exp(x)` for large x. Again, that can very easily lead to numerical instability and `nan`s (if you are lucky) or numerical bugs that are extemely hard to find (if you are unlucky).

To handle this we notice that the softplus function is somewhat similar to the rectified linear unit function (ReLU): $\text{softplus}(t)$ is very closed to zero when $t<<0$ and is very closed to $t$ is $t>>0$.

Note: it does not save any computation time, because all cases are computed anyway. It leads to better numerical stability of an optimization scheme.

In the softplus function, compute_real_mask is a mask matrix that filters all elements in the range [-limit, limit]. The function computes softplus for x in this range using $\text{softplus}(x) = \log(1+\exp(x))$ in the following line. 
> computed_real_part = np.log(1 + np.exp(x*compute_real_mask)*compute_real_mask)

return_same_mask is a mask that filters all elements that is equal or greater than limit. The function computes softplus for x in this range using $x$ as an approximation in the following lines. 

> returned_same_part = x*return_same_mask

> return computed_real_part + returned_same_part

Finally, for elements in $x$ that is equal or smaller than -limit, the function computes softplus for x in this range using $0$ as an approximation, implied in the following line. 

> computed_real_part = np.log(1 + np.exp(x*compute_real_mask)*compute_real_mask)

To test the above implement, the test_softplus function using all integers from -200 to 200 to see the difference between the approximated implement and the real softplus function. The errors should be less than 1e-3 since no errors are prompted.

```python
1e10+1e-10 == 1e10
```

```python
import numpy as np

def softplus(x_, limit=5):
    x = np.array(x_)
    compute_real_mask = np.logical_and(-limit < x, x < limit)
    return_same_mask = x >= limit
    computed_real_part = np.log(1 + np.exp(x*compute_real_mask)*compute_real_mask)
    returned_same_part = x*return_same_mask
    return computed_real_part + returned_same_part

def test_softplus():
    x_arr = np.linspace(-200, 200)
    softplus_true = np.log(1 + np.exp(x_arr))
    softplus_stable = softplus(x_arr)
    assert np.max(abs(softplus_true-softplus_stable)) < 1e-3
    
test_softplus()

def logistic(x):
    return 1/(1+np.exp(-x))
```

**Q5.3**: Write (by filling in the missing code below) Gradient Descent and Stochastic Gradient Descent algorithm for the MAP loss and plot 

- MAP loss
- L2 distance between current theta and true one
- cosine distance between current theta and true one

use results from **Q4.6, Q4.7**. Note that our loss is $loss(\theta) = -\mathcal L(\theta)$ and it is usually referred as "the negative log-likelihood".

```python
from numpy.linalg import norm
from scipy.spatial.distance import cosine
import matplotlib.cm as cm

def logistic_MAP_loss(X, y, theta_current, lambd):
    return (- y.T @ X @ theta_current 
            + np.sum(softplus(X @ theta_current))
            + lambd*norm(theta_current)**2)

def logistic_MAP_loss_gradient(X, y, theta_current, lambd):
    return -X.T @ (y - logistic(X @ theta_current)) + 2 * lambd * theta_current

def logistic_MAP_loss_single_gradient(X, y, smaple_id, theta_current, lambd):
    x_i = X[smaple_id].reshape((1, -1))
    y_i = y[smaple_id].reshape((1, 1))
    return logistic_MAP_loss_gradient(x_i, y_i, theta_current, lambd)

def logistic_classification_accuracy(X, y, theta_current):
    return np.mean(np.equal(y, (X @ theta_current) > 0)) ## ! -- code required

def run_MAP_gd(X, y, lambd, learning_rate = 0.01, max_epoch_n=200, init_theta=None):
    theta = init_theta if init_theta is not None else np.ones((X.shape[1], 1))*5
    steps = []
    for epoch_n in range(max_epoch_n):
        theta = theta - learning_rate * logistic_MAP_loss_gradient(X, y, theta, lambd)
        steps.append(theta)
    return steps

def run_MAP_sgd(X, y, lambd, learning_rate = 0.01, max_epoch_n=200, init_theta=None):
    theta = init_theta if init_theta is not None else np.ones((X.shape[1], 1))*5
    steps = []
    for epoch_n in range(max_epoch_n):
        for smaple_id in range(X.shape[0]):
            ## ! -- code required
            theta = theta - learning_rate * logistic_MAP_loss_single_gradient(X, y, smaple_id, theta, lambd)
        steps.append(theta)
    return steps
```

```python
import time

def draw_gd_plots(axes, theta_steps, X, y, lambd, theta_real):
    collected_val_dicts = []
    for theta in theta_steps:
        collected_val_dicts.append({
                'distance_to_true': np.linalg.norm(theta-theta_real),
                'cosine_dist_to_true': cosine(theta, theta_real),
                'map_loss': float(logistic_MAP_loss(X, y, theta, lambd)),
            })
        
    
    for plot_id, key in enumerate(collected_val_dicts[0].keys()):
        subfig = axes[plot_id]
        subfig.set_title(key)
        subfig.plot(list(d[key] for d in collected_val_dicts))

def main():
    for k_featues in [2, 100]:
        data_args = {'n_samples': 200, 'k_featues': k_featues, 'sigma_eps': 0.3, 
                     'sigma_theta': 5, 'feature_correlation_rate': 0}
        data = generate_logistic_regression_data(**data_args)
        X_train, y_train, X_test, y_test, theta_real = data
        fig, axes = plt.subplots(1, 3, figsize=(10,3))
        lamdb = 0
        
        now = time.clock()
        steps = run_MAP_gd(X_train, y_train, lamdb, learning_rate=1e-3, max_epoch_n=1000)
        draw_gd_plots(axes, steps, X_train, y_train, lamdb, theta_real)
        print('gd runtime:', time.clock() - now)
        
        now = time.clock()
        steps = run_MAP_sgd(X_train, y_train, lamdb, learning_rate=1e-3, max_epoch_n=1000)
        draw_gd_plots(axes, steps, X_train, y_train, lamdb, theta_real)
        fig.suptitle('GD vs SGD, k_featues = %d' % k_featues, verticalalignment='bottom')
        print('sgd runtime:', time.clock() - now)
        
        plt.show()
    
main()
```

```python
from itertools import product

def draw_f_contour(subfig, f, x_limits, n_lines=30, n=100):
    space = [np.linspace(*x_limit, n) for x_limit in x_limits]
    f_vals = np.array([f(np.array(v)) for v in product(*space)]).reshape(n, n).T
    subfig.set_xlim(*x_limits[0])
    subfig.set_ylim(*x_limits[1])
    subfig.contour(*space, f_vals, n_lines)
    
def draw_gd_steps(subfig, steps, func, theta_real):
    around_space = list([-10, 10] for coord in theta_real)
    draw_f_contour(subfig, func, around_space)
    subfig.plot(np.hstack(steps)[0], np.hstack(steps)[1], marker='+', alpha=0.5)
    subfig.scatter(*np.ravel(theta_real), c='r', label='true theta')
    theta_best = steps[-1]
    subfig.scatter(*np.ravel(theta_best), c='g', label='found theta') 

def main():
    data_args = dict(n_samples=200, k_featues=2, sigma_eps=0.1, sigma_theta=5)
    data = generate_logistic_regression_data(**data_args)
    X_train, y_train, X_test, y_test, theta_real = data
    
    lambds = [0, 0.1, 1, 5, 10, 20]
    for lambd, subfig in subfig_iter(lambds, val_name='lambda', title='MAP loss'):
        steps = run_MAP_gd(X_train, y_train, lambd)
        draw_func = lambda theta: logistic_MAP_loss(X_train, y_train, theta, lambd) ## ! -- code required
        draw_gd_steps(subfig, steps, draw_func, theta_real)
    plt.show()

main()
```

**A5.4:**

The contour of the regularization term $\|\theta\|^2$ is of circles. Adding the regularization term to the loss function make the contour of loss function like circles. Or in the other word, decrease the condition number of the Hessian. This will increase the convergent speed of the GD. But a large regularization parameter (i.e. prior that enforces low variance of resulting weight distribution) may dominate the loss function (likelihood term) and makes the found theta deviate from the real one.

For a fixed lambda, as the number of samples increases, it converges to a closer to the true value. This is expected since seeing more data points gives us extra information about the system. However, when lambda is 0, we do not include prior probilities (in our case it is related to the noise distribution), so $\theta$ tends to become bigger(prior was choosing smaller values for theta) When we make lambda=2, we included the prior so it converged to a better value. It is more likely that the gradient descent method converges to a finite point, even when the problem is ill-posed.

**Q5.6**: Using the provided code below, draw several contour plots of logistic loss with different degrees of feature interdependence. Explain your results.

```python
def main():
    rates = np.linspace(0, 0.9, 4)
    for feature_correlation_rate, subfig in subfig_iter(rates, val_name='feature_correlation_rate'):
        data_args = dict(n_samples=1000, k_featues=2, sigma_eps=0.1, sigma_theta=3)
        data_args['feature_correlation_rate'] = feature_correlation_rate
        data = generate_logistic_regression_data(**data_args)
        X_train, y_train, X_test, y_test, theta_real = data
        steps = run_MAP_gd(X_train, y_train, 0)
        draw_func = lambda th: logistic_MAP_loss(X_train, y_train, th, 0)
        draw_gd_steps(subfig, steps, draw_func, theta_real)
    plt.show()

main()
```

**A5.6**

These plots suggest the more feature correlation is, the bigger condition number is. In the other word, correlations in features make the contour of the loss function flats and slow down the convergence of the optimization. This is because the Hessian of the logistic function has form $X^T S X$, where $S$ is a diagonal matrix. (See Ref. below.) This term behaves similarly to the covariance of the samples. If feature correlation is large, the condition number of the Hessian is also large (https://www.cs.ox.ac.uk/people/nando.defreitas/machinelearning/lecture6.pdf)

If we have full correlation, namely $\alpha = 0$ , then all of our features become linear combination of the features, so we need to know $\theta_{1} + \theta_{0}$, not $\theta$ matrix, to make an estimation. This results in linear contour plots, since we get a line equation. So we can observe that as the correlation rate increases, the contours become more linear. And we end up with a less accurate estimate since we loose information with correlation. 

Actually, high correlation case leads to ill-posed problem, i.e. it has multiple equvalent solutions. If we have full correlation, all thetas that satisfy $\theta_1+\theta_2=k$ are equivalent for each $k$. That leads to numerical instability, because there is no single minima to converge - there is an entire subspace of correct solutions.

**Q5.7**: Using the provided code below, draw plots of train\test loss with different parameters. Explain your results.

```python
from collections import OrderedDict

def draw_losses():
    arg_ranges = OrderedDict([
        ('k_featues', np.linspace(10, 300, 40).astype(int)),
        ('n_samples', np.linspace(10, 300, 40).astype(int)),
        ('feature_correlation_rate', np.linspace(0, 1, 50).astype(float)),
        ('sigma_eps', np.linspace(0.01, 5, 50).astype(float)),
    ])

    k_range = [2, 100]
    l2_lambda = 0
    overall_iterator = subfig_iter(list(product(arg_ranges.items(), k_range)), n_col=2)
    for ((arg_name, arg_vals), k_featues_default), subfig in overall_iterator:
        collected_val_dics = []
        for arg_val in arg_vals:
            data_args = dict(n_samples=200, k_featues=k_featues_default, sigma_eps=0.1, 
                             sigma_theta=1, feature_correlation_rate=0, train_test_ratio=4)
            data_args.update({arg_name: arg_val})
            data = generate_logistic_regression_data(**data_args)
            X_train, y_train, X_test, y_test, theta_real = data
            init_theta = np.zeros((data_args['k_featues'], 1))
            
            gd_args = dict(max_epoch_n=100, init_theta=init_theta)
            steps = run_MAP_gd(X_train, y_train, l2_lambda, **gd_args)
            best_theta = steps[-1]
            
            train_acc = logistic_classification_accuracy(X_train, y_train, best_theta)
            test_acc = logistic_classification_accuracy(X_test, y_test, best_theta)
            collected_val_dics.append(
                OrderedDict([
                    ('train_acc', train_acc),
                    ('test_acc', test_acc)
                ]))

        for key, color in zip(collected_val_dics[0].keys(), ['b', 'g']):
            subfig.set_title('%s (k_featues_default = %s)'%(arg_name, k_featues_default))
            track_vals = list(d[key] for d in collected_val_dics)
            subfig.plot(arg_vals, track_vals, marker='.', label=key, color=color, alpha=0.1)
            ws = 7
            assert ws % 2 == 1
            half_ws = int((ws-1)/2)
            track_vals_avg = np.convolve(track_vals, np.ones(ws)/ws, mode='valid')
            plot_args = dict(color=color, label=key+'_avg')
            subfig.plot(arg_vals[half_ws:-half_ws], track_vals_avg, '--', **plot_args)
            subfig.set_ylim(0.4, 1.05)
            subfig.set_ylabel('accuracy')
    plt.show()

draw_losses()
```

**A5.7:**

The first two plots tell the same story. (Actually, only one figure is needed.) They suggest, with the same number of samples, increasing the number of features will increase the train_acc, but will reduce the test_acc. Or in other words, it will overfit the model. The train_acc increases because the signal-noise ratio increase when a number of features increases. The overfitting occurs because a model with more parameters needs more samples, while the number of samples is fixed.

The third and forth figures show the effect of changing the number of samples. When there are two features, the difference can barely see. This is because the model is underfitted. When there are 100 features, the test_acc increases when the number of the samples increases. This is because increasing number of samples is a way to prevent the overfitting of the model.

The fifth and the sixth figures tell what happen when features are correlated. For the two features case. the difference still hardly tell, since it is underfitted. For the 100 features cases, increasing the correlation of the features will deduce the train and test accuracy. We basically reduce effective amount of features (information) that we pass to our model, having number of samples and complexity of a model fixed. Moveover, if you project high-dimentional data into lower dimentions, it becomes harder to separate. And if you look at figure in **Problem 5**, you will see that problem does not have a unique solution if all samples lay in a subspace of original space.

The final two plots show the effect of changing noise variances. Both accuracies decrease when the variances of noise increase. This is due to the increasing of the signal-noise ratio makes it hard to recover the ground truth. However, it seems like high dimentional model can fit even to high noise (train accuracy is still high). Fewer samples in high dim space may become far from each other, so the noise level may not change training data enough to be misclassified. But we still can see its effect on test set since the decision rule may not be correct.
